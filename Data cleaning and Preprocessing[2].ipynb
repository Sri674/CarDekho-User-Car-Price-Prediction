```python
import pandas as pd


def convert_price(price):
    price = price.replace('₹', '').strip()

    price = price.replace(',', '')
    
    # Check for 'Lakh' and 'Crore' and handle accordingly
    if 'Lakh' in price:
        # Remove ' Lakh' and convert to float, then multiply by 100,000
        return float(price.replace(' Lakh', '').strip()) * 100000
    elif 'Crore' in price:
        # Remove ' Crore' and convert to float, then multiply by 10,000,000
        return float(price.replace(' Crore', '').strip()) * 10000000
    else:
        # For prices without 'Lakh' or 'Crore', just convert to float
        return float(price.strip())

# Apply the conversion function to the 'price' column
overall_totalcities_df2['price'] = overall_totalcities_df2['price'].replace({'₹': ''}, regex=True)  
overall_totalcities_df2['price'] = overall_totalcities_df2['price'].apply(convert_price)  

# Convert the 'price' column to integer (if desired, as it seems like whole numbers)
overall_totalcities_df2['price'] = overall_totalcities_df2['price'].astype(int)

print(overall_totalcities_df2['price'])
```

```python
import pandas as pd


def categorize_engine_cc(cc_value):
    if cc_value < 800:
        return " less than 800"
    elif 800 <= cc_value < 1500:
        return "1000"
    elif 1500 <= cc_value <2500:
        return "2000"
    elif 2500 <= cc_value <3500:
        return "3000"
    elif 3500 <= cc_value <4500:
        return "4000"
    elif 4500 <= cc_value < 5500:
        return "5000"
    else:
        return "Above 5000"

overall_totalcities_df21["Engine_category"]= overall_totalcities_df21["Engine"].apply(categorize_engine_cc)
print(overall_totalcities_df21[["Engine", "Engine_category"]])



```

```python
from sklearn.preprocessing import LabelEncoder
import joblib

# List of categorical columns to encode
categorical_columns = ['fuel type', 'transmission', 'Ownership', 'City', 'Engine_category', 'Body type', 'oem']

encoders = {}

# Apply LabelEncoder for each column
for column in categorical_columns:
    # Initialize the LabelEncoder
    le = LabelEncoder()
    
    # Fit the encoder on the column and transform it
    overall_totalcities_df3[column] = le.fit_transform(overall_totalcities_df3[column])
    
    # Save the encoder for future use
    encoders[column] = le

# Save the encoders to a file
joblib.dump(encoders, 'categorical_encoders.pkl')

# Optionally, display the first few rows to verify the result
overall_totalcities_df3.head(10)
```

```python
import numpy as np

# List of columns you want to adjust
columns_to_adjust = ['Kms Driven', 'price', 'Max Power', 'Torque', 'Mileage']

# Loop through each column and apply the np.where logic
for column in columns_to_adjust:
    # Calculate Q1, Q3, and IQR for each column
    Q1 = overall_totalcities_df3[column].quantile(0.25)
    Q3 = overall_totalcities_df3[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper limits for each column
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Apply a check to make sure the lower limit is not negative for relevant columns
    if column in ['Kms Driven', 'price', 'Torque', 'Mileage']:
        lower_limit = max(0, lower_limit)  # Set lower limit to 0 if negative
    
    # Print the lower and upper limits for each column
    print(f"Column: {column}")
    print(f"Lower Limit: {lower_limit}")
    print(f"Upper Limit: {upper_limit}")
    print("-" * 40)
    
    # Apply np.where to handle outliers
    overall_totalcities_df3[column] = np.where(
        overall_totalcities_df3[column] > upper_limit, upper_limit,
        np.where(overall_totalcities_df3[column] < lower_limit, lower_limit, overall_totalcities_df3[column])
    )

# Display the updated DataFrame (first few rows for inspection)
print(overall_totalcities_df3.head())
```

```python
# Applying the IQR method with proper capping
lower_limit = 0
upper_limit = 1776500

# Cap values below the lower limit (0) and above the upper limit
overall_totalcities_df3['price'] = np.where(overall_totalcities_df3['price'] < lower_limit, lower_limit,
                                             np.where(overall_totalcities_df3['price'] > upper_limit, upper_limit,
                                                      overall_totalcities_df3['price']))

# Check the minimum value again
print("Min value after capping:", overall_totalcities_df3['price'].min())
```
